```python
def map_comp(tematica):
    if tematica.count('-')==1:
        componente = tematica[6:tematica.find('-')-1]
    else:
        componente = tematica[6:tematica.find('I')-4]
    return componente

def map_investimento(tematica):
    # con numeri
    if tematica.count('-')==1:
        investimento = tematica[tematica.find('-')+2:]
    else:
        first = tematica.find('-')+1
        tematica = tematica[first:]
        investimento = tematica[tematica.find('-')+2:]
    return investimento

"""def map_investimento(tematica):
    # senza numeri
    investimento = tematica[tematica.find('-')+7:]
    return investimento"""



tmp["CODICE_COMPONENTE"] = tmp["NOME_TEMATICA"].apply(lambda x: x[:4])
tmp["CODICE_MISSIONE"] = tmp["NOME_TEMATICA"].apply(lambda x: x[:2])
tmp["COMPONENTE"] = tmp['NOME_TEMATICA'].apply(map_comp)
tmp["INVESTIMENTO"] = tmp['NOME_TEMATICA'].apply(map_investimento)
```
